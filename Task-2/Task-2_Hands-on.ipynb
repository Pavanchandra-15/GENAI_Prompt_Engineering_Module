## LangChain Hands-on Implementation

This notebook demonstrates practical implementation of LangChain concepts including:

- LLMs and Chat Models  
- Prompt Templates  
- Chains  
- Memory  
- Agents and Tools  

All examples are implemented using Python and LangChain.

```python
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Create pipeline
pipe = pipeline(
    "text-generation",
    model="gpt2"
)

# Wrap with LangChain
llm = HuggingFacePipeline(pipeline=pipe)

# Run query
response = llm.invoke("Explain LangChain in simple words")

print(response)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100,
    temperature=0.7
)

llm = HuggingFacePipeline(pipeline=pipe)

response = llm.invoke(
    "Explain LangChain in simple words"
)

print(response)
```

```python
# Prompt templates 

from langchain.prompts import PromptTemplate

# Create template
template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms in 3 sentences only."
)

# Format prompt
formatted_prompt = template.format(topic="Computer Vision")

# Pass to LLM
response = llm.invoke(formatted_prompt)

print(response)
```

```python
# Chains 

from langchain.chains import LLMChain

# Reuse previous template
chain = LLMChain(
    llm=llm,
    prompt=template
)

# Run chain
response = chain.run("Artificial Intelligence")

print(response)
```

```python
# Memory

from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

# Initialize memory
memory = ConversationBufferMemory()

# Create conversation chain
conversation = ConversationChain(
    llm=llm,
    memory=memory
)

# First interaction
print(conversation.predict(input="Remember: My name is Pavan"))
# Second interaction (should remember name)
print(conversation.predict(input="What is my name?"))
```

```python
# Agents and tools

from langchain.agents import initialize_agent, Tool

# Create a simple calculator tool
def calculator(query):
    return str(eval(query))

tool = Tool(
    name="Calculator",
    func=calculator,
    description="Useful for solving math problems"
)

# Initialize agent
agent = initialize_agent(
    tools=[tool],
    llm=llm,
    agent="zero-shot-react-description",
    verbose=True,
    handle_parsing_errors=True   # Important Fix right here
)

response = agent.run("Calculate 25 * 4?")
print(response)
```

```python
# NOTE:
# Agents may not work properly with lightweight models like GPT-2.
# This is because agents require strong reasoning and structured output.

print("Agent demonstration completed (may not work properly with GPT-2)")
```

###### Agents require instruction-following and reasoning models. Since GPT-2 is a basic generative model, it fails in structured agent workflows. However, the implementation is correct and works with advanced models like GPT-3.5 or GPT-4.
